```python
import pandas as pd
import numpy as np
import os
import statistics as stats
data = pd.read_csv('/home/jovyan/work/datasets/Sensor_readings.csv')
data.drop(columns=['mac', 'boardtype', 'Position', 'model', 'rowid'], inplace=True)
data.head(10)
data.drop(['temp_max', 'temp_min', 'light_max','light_min', 'humidity_min', 'humidity_max'], axis=1, inplace=True)

def median(Series):
    n = len(Series)
    index = n // 2
    if n % 2: 
        return sorted(Series)[index]
    else:
        return sum(sorted(Series)[index - 1:index + 1]) / 2

def print_stats(median,title='Median for locations:',data=data):
    print ('Median for locations:')
    for locationName, locationData in data.groupby('location'):
        print(locationName)
        str1 = '\ttemp\t\t'
        str2 = median(locationData.temp_avg)     
        print(str1, str2)
        str3 = '\tlight\t\t'
        str4 = median(locationData.light_avg)    
        print(str3, str4)
        str5 = '\thumidity\t'
        str6 = median(locationData.humidity_avg)    
        print(str5, str6)

        for boardid, boardData in locationData.groupby('boardid'):
            elevation = boardData['elevation'].dropna().mean()
            print('\t',("board - %s"%(boardid)), ' - ', ("elevation - %s"%(elevation)))
            str7 = '\t\ttemp\t\t'
            str8 = median(boardData.temp_avg)
            print(str7, str8)
            str9 = '\t\tlight\t\t'
            str10 = median(boardData.light_avg)
            print(str9, str10)
            str11 = '\t\thumidity\t'
            str12 = median(boardData.humidity_avg)
            print(str11, str12)

print_stats(median,title='Median for locations:',data=data)
```
